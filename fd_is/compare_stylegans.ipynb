```python
from score_infinity import calculate_FID_infinity_path, calculate_IS_infinity_path
def get_scores(fake_path, min_fake = 500):
    FID_infinity = calculate_FID_infinity_path('output_statistics.npz', fake_path, batch_size=10, min_fake=min_fake)
    IS_infinity = calculate_IS_infinity_path(fake_path, batch_size=10, min_fake=min_fake)
    return FID_infinity, IS_infinity

data_paths = ["stylegan1","stylegan2"]
fid_list, is_list = [],[]
for data in data_paths:
    fid, isi = get_scores(data)
    print(f"For {data} generated dataset, FID_infty={fid:.3f} and IS_infty={isi:.3f}")
    fid_list.append(fid)
    is_list.append(isi)
```

```python
for data in data_paths:
    fid, isi = get_scores(data, 2000)
    print(f"For {data} generated dataset, FID_infty={fid:.3f} and IS_infty={isi:.3f}")
    fid_list.append(fid)
    is_list.append(isi)
```
